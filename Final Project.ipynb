```python
# ════════════════════════════════════════════════════════════════
# CELL 5: CLAN RECRUITMENT MATCHMAKER
# Jalankan SETELAH Cell 1!
# ════════════════════════════════════════════════════════════════

def fmt_r(n): return f"{int(n):,}".replace(',','.')

# ── KETERANGAN WIDGET ─────────────────────────────────────────────────────────
info_box = widgets.HTML("""
<div style='background:linear-gradient(135deg,#0d1a2e,#0a0f1e);
            border:1px solid #4a90d966;border-radius:10px;
            padding:12px 16px;margin-bottom:12px;
            font-family:Exo 2,sans-serif;'>
  <div style='color:#4a90d9;font-size:11px;font-weight:800;
              letter-spacing:1.5px;text-transform:uppercase;margin-bottom:6px;'>
    ℹ️ Tentang Widget Ini
  </div>
  <div style='color:#8aafd4;font-size:11px;line-height:1.7;'>
    Widget ini adalah <b style='color:#f5c518;'>Decision Support System</b> untuk rekrutmen klan.
    Sistem menggunakan <b style='color:#00d1b2;'>Rule-Based Scoring</b> — setiap kandidat
    dinilai berdasarkan kriteria terbobot sesuai tipe klanmu.<br><br>
    <b style='color:#b94fff;'>Cara pakai:</b> Pilih tipe klan → atur standar minimum →
    input data kandidat → sistem otomatis hitung skor & beri verdict.
  </div>
</div>""")

# ── TIPE KLAN & BOBOT ─────────────────────────────────────────────────────────
CLAN_PROFILES = {
    'War Clan':         {'th':25,'donate':15,'war_wr':30,'activity':10,'trophy':10,'hero':10,
                         'desc':'Fokus CWL & Clan War. Prioritas win rate & hero level tinggi.'},
    'Farming Clan':     {'th':20,'donate':30,'war_wr':10,'activity':25,'trophy':5,'hero':10,
                         'desc':'Fokus loot & donasi. Prioritas aktifitas & donation rate.'},
    'Casual Clan':      {'th':15,'donate':20,'war_wr':15,'activity':30,'trophy':10,'hero':10,
                         'desc':'Santai & ramah pemula. Aktifitas lebih penting dari stats.'},
    'Competitive Clan': {'th':25,'donate':15,'war_wr':25,'activity':15,'trophy':10,'hero':10,
                         'desc':'Semua kriteria ketat. Hanya untuk player berpengalaman.'},
}

dd_clan_type = widgets.Dropdown(
    options=list(CLAN_PROFILES.keys()),
    value='War Clan',
    description='Tipe Klan:',
    style={'description_width':'90px'},
    layout=widgets.Layout(width='250px'))

clan_desc_out = widgets.Output()

def update_clan_desc(change=None):
    profile = CLAN_PROFILES[dd_clan_type.value]
    w = profile
    with clan_desc_out:
        clear_output(wait=True)
        display(widgets.HTML(f"""
        <div style='background:#0d1520;border:1px solid #2a3a50;border-radius:8px;
                    padding:10px 14px;font-family:Exo 2,sans-serif;margin-top:6px;'>
          <div style='color:#8aafd4;font-size:10px;margin-bottom:6px;'>{w['desc']}</div>
          <div style='display:flex;flex-wrap:wrap;gap:6px;'>
            <span style='background:#1a2a40;border-radius:4px;padding:3px 8px;
                         font-size:9px;color:#f5c518;'>TH: {w['th']}%</span>
            <span style='background:#1a2a40;border-radius:4px;padding:3px 8px;
                         font-size:9px;color:#00d1b2;'>Donate: {w['donate']}%</span>
            <span style='background:#1a2a40;border-radius:4px;padding:3px 8px;
                         font-size:9px;color:#e74c3c;'>War WR: {w['war_wr']}%</span>
            <span style='background:#1a2a40;border-radius:4px;padding:3px 8px;
                         font-size:9px;color:#f5a623;'>Aktif: {w['activity']}%</span>
            <span style='background:#1a2a40;border-radius:4px;padding:3px 8px;
                         font-size:9px;color:#b94fff;'>Trophy: {w['trophy']}%</span>
            <span style='background:#1a2a40;border-radius:4px;padding:3px 8px;
                         font-size:9px;color:#4a90d9;'>Hero: {w['hero']}%</span>
          </div>
        </div>"""))

dd_clan_type.observe(update_clan_desc, names='value')
update_clan_desc()

# ── STANDAR MINIMUM KLAN ──────────────────────────────────────────────────────
min_th = widgets.IntSlider(
    value=10, min=1, max=16,
    description='Min TH:',
    style={'description_width':'100px','handle_color':'#f5c518'},
    layout=widgets.Layout(width='320px'))

min_trophy = widgets.IntSlider(
    value=1500, min=0, max=6000, step=100,
    description='Min Trophy:',
    style={'description_width':'100px','handle_color':'#b94fff'},
    layout=widgets.Layout(width='320px'))

min_war_wr = widgets.IntSlider(
    value=50, min=0, max=100,
    description='Min War WR%:',
    style={'description_width':'100px','handle_color':'#e74c3c'},
    layout=widgets.Layout(width='320px'))

# ── INPUT KANDIDAT ────────────────────────────────────────────────────────────
cand_name = widgets.Text(
    value='', placeholder='Nama player...',
    description='Nama:',
    style={'description_width':'100px'},
    layout=widgets.Layout(width='300px'))

cand_th = widgets.Dropdown(
    options=list(range(1,17)),
    value=10,
    description='Town Hall:',
    style={'description_width':'100px'},
    layout=widgets.Layout(width='200px'))

cand_trophy = widgets.BoundedIntText(
    value=2000, min=0, max=9999,
    description='Trophy:',
    style={'description_width':'100px'},
    layout=widgets.Layout(width='200px'))

cand_donate = widgets.BoundedIntText(
    value=500, min=0, max=99999,
    description='Donasi/Season:',
    style={'description_width':'100px'},
    layout=widgets.Layout(width='200px'))

cand_war_wr = widgets.BoundedIntText(
    value=60, min=0, max=100,
    description='War Win Rate%:',
    style={'description_width':'100px'},
    layout=widgets.Layout(width='200px'))

cand_war_stars = widgets.BoundedIntText(
    value=2, min=0, max=3,
    description='Avg War Stars:',
    style={'description_width':'100px'},
    layout=widgets.Layout(width='200px'))

cand_activity = widgets.Dropdown(
    options=[('Sangat Aktif (Daily)',4),('Aktif (4-5x/minggu)',3),
             ('Sedang (2-3x/minggu)',2),('Jarang (1x/minggu)',1)],
    value=3,
    description='Aktivitas:',
    style={'description_width':'100px'},
    layout=widgets.Layout(width='280px'))

cand_hero_avg = widgets.BoundedIntText(
    value=50, min=0, max=100,
    description='Avg Hero Level:',
    style={'description_width':'100px'},
    layout=widgets.Layout(width='200px'))

cand_lang = widgets.Dropdown(
    options=['Indonesia 🇮🇩','English 🌐','Bilingual 🌏'],
    value='Indonesia 🇮🇩',
    description='Bahasa:',
    style={'description_width':'100px'},
    layout=widgets.Layout(width='220px'))

btn_check = widgets.Button(
    description='🎯 CEK KANDIDAT',
    layout=widgets.Layout(width='100%', height='42px', margin='12px 0 4px'))
btn_check.style.button_color = '#7b00cc'
btn_check.style.font_weight  = 'bold'

btn_reset_r = widgets.Button(
    description='↺ Reset',
    layout=widgets.Layout(width='100%', height='32px'))
btn_reset_r.style.button_color = '#1a0828'
btn_reset_r.style.font_weight  = 'bold'

out_recruit_result = widgets.Output()

# ── SCORING ENGINE ────────────────────────────────────────────────────────────
def on_check(b):
    profile = CLAN_PROFILES[dd_clan_type.value]
    w = profile

    # Hard filter dulu
    hard_fail = []
    if cand_th.value < min_th.value:
        hard_fail.append(f"TH terlalu rendah (TH{cand_th.value} < min TH{min_th.value})")
    if cand_trophy.value < min_trophy.value:
        hard_fail.append(f"Trophy kurang ({cand_trophy.value} < {min_trophy.value})")
    if cand_war_wr.value < min_war_wr.value:
        hard_fail.append(f"War Win Rate kurang ({cand_war_wr.value}% < {min_war_wr.value}%)")

    # Scoring (0-100 per kriteria)
    th_score      = min(100, (cand_th.value / 16) * 100)
    donate_score  = min(100, (cand_donate.value / 2000) * 100)
    war_wr_score  = cand_war_wr.value
    activity_score= (cand_activity.value / 4) * 100
    trophy_score  = min(100, (cand_trophy.value / 6000) * 100)
    hero_score    = min(100, cand_hero_avg.value)

    total = (
        th_score      * w['th']      / 100 +
        donate_score  * w['donate']  / 100 +
        war_wr_score  * w['war_wr']  / 100 +
        activity_score* w['activity']/ 100 +
        trophy_score  * w['trophy']  / 100 +
        hero_score    * w['hero']    / 100
    )

    # Verdict
    if hard_fail:
        verdict, vc, vi = 'TIDAK MEMENUHI SYARAT', '#e74c3c', '❌'
    elif total >= 80:
        verdict, vc, vi = 'SANGAT DIREKOMENDASIKAN', '#f5c518', '🏆'
    elif total >= 65:
        verdict, vc, vi = 'DIREKOMENDASIKAN', '#27ae60', '✅'
    elif total >= 50:
        verdict, vc, vi = 'PERTIMBANGKAN', '#f5a623', '⚠️'
    else:
        verdict, vc, vi = 'TIDAK DIREKOMENDASIKAN', '#e74c3c', '❌'

    name = cand_name.value or 'Unknown Player'

    def score_bar(score, color):
        return (f"<div style='background:#0a0410;border-radius:50px;height:6px;"
                f"overflow:hidden;margin-top:3px;'>"
                f"<div style='background:{color};width:{score:.0f}%;height:100%;"
                f"border-radius:50px;box-shadow:0 0 6px {color};'></div></div>")

    criteria_rows = [
        ('🏰 Town Hall',    th_score,       w['th'],      '#f5c518'),
        ('📦 Donasi',       donate_score,   w['donate'],  '#00d1b2'),
        ('⚔️ War WR',      war_wr_score,   w['war_wr'],  '#e74c3c'),
        ('🟢 Aktivitas',    activity_score, w['activity'],'#27ae60'),
        ('🏆 Trophy',       trophy_score,   w['trophy'],  '#b94fff'),
        ('⚡ Hero Level',   hero_score,     w['hero'],    '#4a90d9'),
    ]

    rows_html = ''.join(
        f"<div style='margin-bottom:10px;'>"
        f"<div style='display:flex;justify-content:space-between;font-size:10px;'>"
        f"<span style='color:#c8a060;'>{lbl}</span>"
        f"<span style='color:{col};font-weight:800;'>{score:.0f}/100"
        f" <span style='color:#3a2a4a;font-size:9px;'>(bobot {bbt}%)</span></span>"
        f"</div>"
        f"{score_bar(score, col)}"
        f"</div>"
        for lbl, score, bbt, col in criteria_rows
    )

    fail_html = ''.join(
        f"<div style='color:#e74c3c;font-size:10px;padding:3px 0;'>⛔ {f}</div>"
        for f in hard_fail
    ) if hard_fail else ""

    with out_recruit_result:
        clear_output(wait=True)
        display(widgets.HTML(f"""
        <div style='font-family:Exo 2,sans-serif;'>

          <!-- Verdict Card -->
          <div style='background:{vc}18;border:2px solid {vc}66;border-radius:12px;
                      padding:16px;text-align:center;margin-bottom:10px;'>
            <div style='font-size:28px;margin-bottom:4px;'>{vi}</div>
            <div style='font-family:Cinzel,serif;color:{vc};font-size:15px;
                        font-weight:900;letter-spacing:2px;'>{verdict}</div>
            <div style='color:#aaa;font-size:12px;margin-top:4px;'>
              {name} · TH{cand_th.value} · {cand_trophy.value} 🏆
            </div>
            <div style='margin-top:10px;font-size:24px;font-weight:900;color:#fff;'>
              SKOR: {total:.1f} / 100
            </div>
          </div>

          <!-- Hard Fails -->
          {('<div style="background:#1a0808;border:1px solid #e74c3c44;border-radius:8px;padding:10px;margin-bottom:10px;">' + fail_html + '</div>') if hard_fail else ''}

          <!-- Breakdown -->
          <div style='background:#0d0515;border:1px solid #2a1040;
                      border-radius:10px;padding:14px;margin-bottom:10px;'>
            <div style='color:#7a5a9a;font-size:9px;font-weight:800;
                        text-transform:uppercase;letter-spacing:1.5px;margin-bottom:10px;'>
              📊 Breakdown Skor ({dd_clan_type.value})
            </div>
            {rows_html}
          </div>

          <!-- Info tambahan -->
          <div style='background:#0d0515;border:1px solid #2a1040;
                      border-radius:8px;padding:10px;font-size:10px;color:#7a5a9a;'>
            <b style='color:#f5a623;'>📋 Info Kandidat:</b><br>
            Bahasa: {cand_lang.value} &nbsp;|&nbsp;
            Avg War Stars: {'⭐'*cand_war_stars.value or 'N/A'} &nbsp;|&nbsp;
            Donasi: {fmt_r(cand_donate.value)}/season
          </div>
        </div>"""))

def on_reset_r(b):
    cand_name.value = ''
    cand_th.value = 10; cand_trophy.value = 2000
    cand_donate.value = 500; cand_war_wr.value = 60
    cand_war_stars.value = 2; cand_activity.value = 3
    cand_hero_avg.value = 50
    with out_recruit_result:
        clear_output()

btn_check.on_click(on_check)
btn_reset_r.on_click(on_reset_r)

# ── ASSEMBLE ──────────────────────────────────────────────────────────────────
recruit_ui = widgets.VBox([
    widgets.HTML("""
    <div style='font-family:Cinzel,serif;color:#f5c518;font-size:14px;
                text-align:center;padding:12px;letter-spacing:3px;
                text-shadow:0 0 12px rgba(245,197,24,0.6);
                background:linear-gradient(135deg,#1a0828,#0d0515);
                border-bottom:2px solid #2a1040;margin-bottom:12px;'>
      🎯 CLAN RECRUITMENT MATCHMAKER
    </div>"""),

    info_box,

    # Klan Config
    widgets.VBox([
        widgets.HTML("<div style='color:#4a90d9;font-size:9px;font-weight:800;"
                     "text-transform:uppercase;letter-spacing:1.5px;margin-bottom:8px;"
                     "font-family:Exo 2,sans-serif;'>🏰 Profil Klan</div>"),
        dd_clan_type, clan_desc_out,
        widgets.HTML("<div style='color:#7a5a9a;font-size:9px;margin:10px 0 4px;"
                     "font-family:Exo 2,sans-serif;'>📏 Standar Minimum</div>"),
        min_th, min_trophy, min_war_wr,
    ], layout=widgets.Layout(
        border='1px solid #4a90d944', border_radius='12px',
        padding='14px', margin='0 0 10px', background_color='#0d0515')),

    # Kandidat Input
    widgets.VBox([
        widgets.HTML("<div style='color:#b94fff;font-size:9px;font-weight:800;"
                     "text-transform:uppercase;letter-spacing:1.5px;margin-bottom:8px;"
                     "font-family:Exo 2,sans-serif;'>🪖 Data Kandidat</div>"),
        cand_name, cand_th, cand_trophy, cand_donate,
        cand_war_wr, cand_war_stars, cand_activity, cand_hero_avg, cand_lang,
        btn_check, btn_reset_r,
    ], layout=widgets.Layout(
        border='1px solid #b94fff44', border_radius='12px',
        padding='14px', margin='0 0 10px', background_color='#0d0515')),

    out_recruit_result,
], layout=widgets.Layout(width='100%', background_color='#0a0608'))

with out_recruit:
    clear_output(wait=True)
    display(recruit_ui)

print("✅ Cell 5 Recruitment Matchmaker siap!")
```

```python
# ════════════════════════════════════════════════════════════════
# CELL 6: MAGIC ITEMS VALUE ANALYSIS
# Jalankan SETELAH Cell 1!
# ════════════════════════════════════════════════════════════════

def fmt_m(n): return f"{int(n):,}".replace(',','.')

# ── KETERANGAN WIDGET ─────────────────────────────────────────────────────────
magic_info = widgets.HTML("""
<div style='background:linear-gradient(135deg,#1a0d2e,#0d0a1e);
            border:1px solid #f5c51844;border-radius:10px;
            padding:12px 16px;margin-bottom:12px;
            font-family:Exo 2,sans-serif;'>
  <div style='color:#f5c518;font-size:11px;font-weight:800;
              letter-spacing:1.5px;text-transform:uppercase;margin-bottom:6px;'>
    ℹ️ Tentang Widget Ini
  </div>
  <div style='color:#c8a060;font-size:11px;line-height:1.7;'>
    Widget ini adalah <b style='color:#f5c518;'>Cost-Benefit Analysis</b> untuk Magic Items.
    Sistem menghitung <b style='color:#00d1b2;'>nilai efisiensi</b> penggunaan item langka
    seperti Book of Building, Hammer of Building, dll.<br><br>
    <b style='color:#b94fff;'>Cara pakai:</b> Pilih Magic Item → input durasi & biaya upgrade
    yang ingin di-skip → sistem hitung apakah item tersebut worth dipakai sekarang
    atau lebih baik disimpan.
  </div>
</div>""")

# ── DATABASE MAGIC ITEMS ──────────────────────────────────────────────────────
MAGIC_ITEMS = {
    # Name: (kategori, deskripsi, nilai_jam_setara, resource_skip)
    'Book of Building':    ('Book',   'Skip 1 upgrade bangunan (max 7 hari)',    168, 'gold'),
    'Book of Heroes':      ('Book',   'Skip 1 upgrade hero (max 7 hari)',         168, 'dark'),
    'Book of Fighting':    ('Book',   'Skip 1 upgrade troops (max 7 hari)',       168, 'elixir'),
    'Book of Spells':      ('Book',   'Skip 1 upgrade spell (max 7 hari)',        168, 'elixir'),
    'Book of Everything':  ('Book',   'Skip upgrade apapun (max 14 hari)',         336, 'any'),
    'Hammer of Building':  ('Hammer', 'Skip upgrade bangunan TANPA resource',     999, 'gold'),
    'Hammer of Heroes':    ('Hammer', 'Skip upgrade hero TANPA resource',         999, 'dark'),
    'Hammer of Fighting':  ('Hammer', 'Skip upgrade troops TANPA resource',       999, 'elixir'),
    'Hammer of Spells':    ('Hammer', 'Skip upgrade spell TANPA resource',        999, 'elixir'),
    'Research Potion':     ('Potion', 'Lab 24x lebih cepat selama 1 jam',          24, 'elixir'),
    'Builder Potion':      ('Potion', 'Semua builder 10x lebih cepat selama 1 jam',10, 'gold'),
    'Training Potion':     ('Potion', 'Instant training troops selama 1 jam',       1, 'elixir'),
    'Resource Potion':     ('Potion', 'Boost resource production 24 jam',          24, 'any'),
    'Power Potion':        ('Potion', 'Max troops level selama 1 jam',              1, 'elixir'),
    'Wall Ring':           ('Ring',   'Setara 1M gold/elixir untuk wall',           0, 'gold'),
    'Hero Equipment (Common)':   ('Equipment','Upgrade hero equipment - Common',   0, 'dark'),
    'Hero Equipment (Epic)':     ('Equipment','Upgrade hero equipment - Epic',     0, 'dark'),
}

CATEGORY_COLORS = {
    'Book':      '#f5c518',
    'Hammer':    '#e74c3c',
    'Potion':    '#00d1b2',
    'Ring':      '#b94fff',
    'Equipment': '#4a90d9',
}

# ── WIDGETS ───────────────────────────────────────────────────────────────────
dd_item = widgets.Dropdown(
    options=list(MAGIC_ITEMS.keys()),
    value='Book of Building',
    description='Magic Item:',
    style={'description_width':'100px'},
    layout=widgets.Layout(width='300px'))

item_desc_out = widgets.Output()

def update_item_desc(change=None):
    cat, desc, val, res = MAGIC_ITEMS[dd_item.value]
    col = CATEGORY_COLORS.get(cat, '#aaa')
    with item_desc_out:
        clear_output(wait=True)
        display(widgets.HTML(f"""
        <div style='background:#0d0515;border:1px solid {col}44;border-radius:8px;
                    padding:8px 12px;margin-top:6px;font-family:Exo 2,sans-serif;'>
          <span style='background:{col}33;border:1px solid {col}66;border-radius:4px;
                       padding:2px 8px;font-size:9px;color:{col};'>{cat}</span>
          <span style='color:#aaa;font-size:10px;margin-left:8px;'>{desc}</span>
        </div>"""))

dd_item.observe(update_item_desc, names='value')
update_item_desc()

# ── INPUT UPGRADE TARGET ──────────────────────────────────────────────────────
inp_upgrade_dur = widgets.IntSlider(
    value=72, min=1, max=336, step=1,
    description='Durasi (jam):',
    style={'description_width':'110px','handle_color':'#f5c518'},
    layout=widgets.Layout(width='350px'))

inp_upgrade_cost_gold = widgets.BoundedIntText(
    value=5000000, min=0, max=999999999, step=100000,
    description='Biaya Gold:',
    style={'description_width':'110px'},
    layout=widgets.Layout(width='240px'))

inp_upgrade_cost_elixir = widgets.BoundedIntText(
    value=0, min=0, max=999999999, step=100000,
    description='Biaya Elixir:',
    style={'description_width':'110px'},
    layout=widgets.Layout(width='240px'))

inp_upgrade_cost_dark = widgets.BoundedIntText(
    value=0, min=0, max=9999999, step=1000,
    description='Biaya Dark:',
    style={'description_width':'110px'},
    layout=widgets.Layout(width='240px'))

# ── KONTEKS SITUASI ───────────────────────────────────────────────────────────
dd_situation = widgets.Dropdown(
    options=[
        ('CWL/War Segera (2-3 hari)', 'war_soon'),
        ('Lagi Rush Upgrade', 'rushing'),
        ('Builder Nganggur', 'idle'),
        ('Upgrade Mahal Banget', 'expensive'),
        ('Santai, tidak ada urgensi', 'casual'),
    ],
    value='casual',
    description='Situasi:',
    style={'description_width':'110px'},
    layout=widgets.Layout(width='300px'))

active_builders = widgets.IntSlider(
    value=2, min=0, max=6,
    description='Builder Aktif:',
    style={'description_width':'110px','handle_color':'#00d1b2'},
    layout=widgets.Layout(width='300px'))

btn_analyze = widgets.Button(
    description='✨ ANALISIS ITEM',
    layout=widgets.Layout(width='100%', height='42px', margin='12px 0 4px'))
btn_analyze.style.button_color = '#4a2a00'
btn_analyze.style.font_weight  = 'bold'

btn_reset_m = widgets.Button(
    description='↺ Reset',
    layout=widgets.Layout(width='100%', height='32px'))
btn_reset_m.style.button_color = '#1a0828'
btn_reset_m.style.font_weight  = 'bold'

out_magic_result = widgets.Output()

# ── ANALISIS ENGINE ───────────────────────────────────────────────────────────
def on_analyze(b):
    item_name = dd_item.value
    cat, desc, max_val_jam, res_type = MAGIC_ITEMS[item_name]
    col   = CATEGORY_COLORS.get(cat, '#aaa')
    dur   = inp_upgrade_dur.value
    gold  = inp_upgrade_cost_gold.value
    elixir= inp_upgrade_cost_elixir.value
    dark  = inp_upgrade_cost_dark.value
    total_cost = gold + elixir + (dark * 200)
    situation = dd_situation.value
    builders  = active_builders.value

    # ── Nilai Item ──────────────────────────────────────────────────────────
    # Efisiensi waktu: seberapa banyak jam yang di-skip vs max kapasitas item
    if cat == 'Book':
        time_eff = min(100, (dur / 168) * 100)
        time_saved = min(dur, 168)
    elif cat == 'Hammer':
        time_eff = 100  # Hammer selalu worth karena skip resource juga
        time_saved = dur
    elif cat == 'Potion':
        if 'Builder' in item_name:
            time_eff = min(100, builders * 20)  # Makin banyak builder aktif makin worth
            time_saved = builders * 10  # 10x speed, 1 jam → efektif N*10 jam
        elif 'Research' in item_name:
            time_eff = 80  # Lab selalu jalan
            time_saved = 24
        else:
            time_eff = 50
            time_saved = 1
    else:
        time_eff = 60
        time_saved = 0

    # ── Resource Efficiency (khusus Hammer) ───────────────────────────────
    resource_skip = total_cost if cat == 'Hammer' else 0
    resource_eff  = min(100, (total_cost / 10000000) * 100) if cat == 'Hammer' else 0

    # ── Situation Modifier ─────────────────────────────────────────────────
    sit_bonus = {
        'war_soon': 25 if cat in ['Book','Hammer'] else 10,
        'rushing':  20,
        'idle':     -20 if cat == 'Builder Potion' else 0,
        'expensive': 20 if cat == 'Hammer' else 5,
        'casual':   -10,
    }.get(situation, 0)

    # ── Final Score ────────────────────────────────────────────────────────
    base_score = time_eff * 0.6 + resource_eff * 0.4
    final_score = min(100, max(0, base_score + sit_bonus))

    if final_score >= 85:
        verdict, vc, vi = 'SANGAT WORTH — PAKAI SEKARANG!', '#f5c518', '🏆'
        rec = 'Kondisi sangat ideal. Jangan tunda lagi!'
    elif final_score >= 70:
        verdict, vc, vi = 'WORTH DIGUNAKAN', '#27ae60', '✅'
        rec = 'Situasi mendukung. Gunakan untuk upgrade ini.'
    elif final_score >= 50:
        verdict, vc, vi = 'PERTIMBANGKAN', '#f5a623', '⚠️'
        rec = 'Boleh dipakai, tapi cari upgrade yang lebih mahal/lama dulu.'
    elif final_score >= 30:
        verdict, vc, vi = 'KURANG WORTH', '#e74c3c', '❌'
        rec = 'Simpan untuk upgrade yang lebih penting.'
    else:
        verdict, vc, vi = 'JANGAN DIPAKAI SEKARANG', '#c0392b', '💀'
        rec = 'Sangat sayang dibuang di sini. Tunggu upgrade TH atau hero.'

    # Durasi display
    def dur_str(h):
        d = h // 24; hr = h % 24
        return (f"{d}h " if d else "") + (f"{hr}j" if hr else "0j")

    with out_magic_result:
        clear_output(wait=True)
        display(widgets.HTML(f"""
        <div style='font-family:Exo 2,sans-serif;'>

          <!-- Verdict -->
          <div style='background:{vc}15;border:2px solid {vc}55;border-radius:12px;
                      padding:16px;text-align:center;margin-bottom:10px;'>
            <div style='font-size:28px;'>{vi}</div>
            <div style='font-family:Cinzel,serif;color:{vc};font-size:14px;
                        font-weight:900;letter-spacing:1.5px;margin-top:4px;'>{verdict}</div>
            <div style='font-size:26px;font-weight:900;color:#fff;margin:8px 0;'>
              {final_score:.0f} / 100
            </div>
            <div style='color:#aaa;font-size:10px;'>{rec}</div>
          </div>

          <!-- Detail Grid -->
          <div style='display:grid;grid-template-columns:1fr 1fr;gap:8px;margin-bottom:10px;'>

            <div style='background:#0d0515;border:1px solid #2a1040;border-radius:8px;padding:10px;'>
              <div style='color:#7a5a9a;font-size:8px;text-transform:uppercase;margin-bottom:6px;'>Waktu Dihemat</div>
              <div style='color:#00d1b2;font-size:18px;font-weight:900;'>{dur_str(time_saved)}</div>
              <div style='color:#3a2a4a;font-size:8px;'>dari {dur_str(dur)} durasi upgrade</div>
            </div>

            <div style='background:#0d0515;border:1px solid #2a1040;border-radius:8px;padding:10px;'>
              <div style='color:#7a5a9a;font-size:8px;text-transform:uppercase;margin-bottom:6px;'>Efisiensi Waktu</div>
              <div style='color:#f5c518;font-size:18px;font-weight:900;'>{time_eff:.0f}%</div>
              <div style='background:#0a0410;border-radius:50px;height:4px;margin-top:4px;'>
                <div style='background:#f5c518;width:{time_eff:.0f}%;height:100%;border-radius:50px;'></div>
              </div>
            </div>

            <div style='background:#0d0515;border:1px solid #2a1040;border-radius:8px;padding:10px;'>
              <div style='color:#7a5a9a;font-size:8px;text-transform:uppercase;margin-bottom:6px;'>Resource Skip</div>
              <div style='color:#e74c3c;font-size:14px;font-weight:900;'>
                {'🥇 '+fmt_m(gold) if gold else ''} {'⚡ '+fmt_m(elixir) if elixir else ''} {'🌑 '+fmt_m(dark) if dark else '–'}
              </div>
            </div>

            <div style='background:#0d0515;border:1px solid #2a1040;border-radius:8px;padding:10px;'>
              <div style='color:#7a5a9a;font-size:8px;text-transform:uppercase;margin-bottom:6px;'>Bonus Situasi</div>
              <div style='color:{"#27ae60" if sit_bonus >= 0 else "#e74c3c"};font-size:18px;font-weight:900;'>
                {"+" if sit_bonus >= 0 else ""}{sit_bonus} pts
              </div>
            </div>
          </div>

          <!-- Item Info -->
          <div style='background:#0d0515;border:1px solid {col}33;
                      border-radius:8px;padding:12px;margin-bottom:8px;'>
            <div style='color:{col};font-size:9px;font-weight:800;
                        text-transform:uppercase;letter-spacing:1.5px;margin-bottom:6px;'>
              📦 {item_name}
            </div>
            <div style='color:#8a7a9a;font-size:10px;line-height:1.6;'>{desc}</div>
            {'<div style="color:#4a90d9;font-size:9px;margin-top:6px;">💡 Builder aktif: '+str(builders)+'x → Potion efektif '+dur_str(builders*10)+'</div>' if 'Builder' in item_name else ''}
          </div>

          <!-- Saran Simpan -->
          <div style='background:#0a0608;border:1px solid #2a1040;
                      border-radius:8px;padding:10px;font-size:10px;color:#5a4a6a;'>
            <b style='color:#f5a623;'>💡 Kapan sebaiknya dipakai:</b><br>
            {"Book: Paling worth untuk upgrade 5-7 hari (mendekati batas 168 jam)." if cat=='Book' else
             "Hammer: Selalu worth untuk upgrade mahal. Prioritaskan TH, Hero, atau Lab." if cat=='Hammer' else
             "Builder Potion: Makin banyak builder aktif, makin worth." if 'Builder' in item_name else
             "Research Potion: Pakai saat lab sedang upgrade lama (Electro Dragon, Titan, dll)." if 'Research' in item_name else
             "Simpan untuk momen yang paling menguntungkan."}
          </div>
        </div>"""))

def on_reset_m(b):
    inp_upgrade_dur.value = 72
    inp_upgrade_cost_gold.value = 5000000
    inp_upgrade_cost_elixir.value = 0
    inp_upgrade_cost_dark.value = 0
    active_builders.value = 2
    with out_magic_result:
        clear_output()

btn_analyze.on_click(on_analyze)
btn_reset_m.on_click(on_reset_m)

# ── ASSEMBLE ──────────────────────────────────────────────────────────────────
magic_ui = widgets.VBox([
    widgets.HTML("""
    <div style='font-family:Cinzel,serif;color:#f5c518;font-size:14px;
                text-align:center;padding:12px;letter-spacing:3px;
                text-shadow:0 0 12px rgba(245,197,24,0.6);
                background:linear-gradient(135deg,#1a0828,#0d0515);
                border-bottom:2px solid #2a1040;margin-bottom:12px;'>
      ✨ MAGIC ITEMS VALUE ANALYSIS
    </div>"""),

    magic_info,

    # Item Selection
    widgets.VBox([
        widgets.HTML("<div style='color:#f5c518;font-size:9px;font-weight:800;"
                     "text-transform:uppercase;letter-spacing:1.5px;margin-bottom:8px;"
                     "font-family:Exo 2,sans-serif;'>📦 Pilih Magic Item</div>"),
        dd_item, item_desc_out,
    ], layout=widgets.Layout(
        border='1px solid #f5c51844', border_radius='12px',
        padding='14px', margin='0 0 10px', background_color='#0d0515')),

    # Upgrade Target
    widgets.VBox([
        widgets.HTML("<div style='color:#00d1b2;font-size:9px;font-weight:800;"
                     "text-transform:uppercase;letter-spacing:1.5px;margin-bottom:8px;"
                     "font-family:Exo 2,sans-serif;'>🏗️ Target Upgrade</div>"),
        inp_upgrade_dur,
        inp_upgrade_cost_gold, inp_upgrade_cost_elixir, inp_upgrade_cost_dark,
    ], layout=widgets.Layout(
        border='1px solid #00d1b244', border_radius='12px',
        padding='14px', margin='0 0 10px', background_color='#0d0515')),

    # Konteks
    widgets.VBox([
        widgets.HTML("<div style='color:#b94fff;font-size:9px;font-weight:800;"
                     "text-transform:uppercase;letter-spacing:1.5px;margin-bottom:8px;"
                     "font-family:Exo 2,sans-serif;'>🎯 Konteks Situasi</div>"),
        dd_situation, active_builders,
        btn_analyze, btn_reset_m,
    ], layout=widgets.Layout(
        border='1px solid #b94fff44', border_radius='12px',
        padding='14px', margin='0 0 10px', background_color='#0d0515')),

    out_magic_result,
], layout=widgets.Layout(width='100%', background_color='#0a0608'))

with out_magic:
    clear_output(wait=True)
    display(magic_ui)

print("✅ Cell 6 Magic Items Analysis siap!")
print("   Dashboard COC sekarang LENGKAP — semua 6 widget aktif! 🎉")
```
